## 08-Second Order Differential Equation - 02

```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ode_func(x, t): 
    # x = [displacement, velocity]
    # x[0] = displacement
    # x[1] = velocity
    
    # dxdt = [d(displacement)/dt, d(velocity)dt]
    # dxdt = [v, d(v)dt]
    

    dxdt = [x[1], -x[0] - x[1]]
    return dxdt
```

```python
# Initial conditions
x0 = [0.0, 1.0]  # Initial displacement and velocity
# Time
t = np.linspace(0, 10, 1000)
```

```python
# Solve the system of differential equations using odeint
solution = odeint(ode_func, x0, t)
```

```python
# Display the solution
plt.plot(t, solution[:, 0], label='Displacement (x)')
plt.plot(t, solution[:, 1], label='Velocity (v)')
plt.xlabel('Time (t)')
plt.ylabel('Displacement and Velocity')
plt.legend()
plt.show()
```

```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ode_func(x, t, k, m):
    # x = [displacement, velocity]
    # x[0] = displacement
    # x[1] = velocity
    
    # dxdt = [d(displacement)/dt, d(velocity)/dt]
    # dxdt = [v, dv/dt]
    
    dxdt = [x[1], -(k/m) * x[0]]
    return dxdt
```

```python
# Initial conditions 1
x01 = [0.1, 0.0]  # Initial displacement and velocity
# k01 = 3675  # N-S
k01 = 18850  # E-W
m01 = 9000

# Initial conditions 2
x02 = [0.2, 0.0]  # Initial displacement and velocity
# k02 = 3675  # N-S
k02 = 18850  # E-W
m02 = 9000

# Time
t = np.linspace(0, 30, 1000)
```

```python
# Solve the system of differential equations using odeint
solution1 = odeint(ode_func, x01, t, args=(k01, m01))
solution2 = odeint(ode_func, x02, t, args=(k02, m02))
```

```python
# Display the solution
plt.plot(t, solution1[:, 0], label='Displacement 1 (x)')
plt.plot(t, solution2[:, 0], label='Displacement 2 (x)')
plt.xlabel('Time (t)')
plt.ylabel('Displacement and Velocity')
plt.legend()
plt.show()
```

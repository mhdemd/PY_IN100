## Maximum Amplitude of Motion in Undamped Free Vibration

```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ode_func(x, t, k, m):
    # x = [displacement, velocity]
    # x[0] = displacement
    # x[1] = velocity
    
    # dxdt = [d(displacement)/dt, d(velocity)/dt]
    # dxdt = [v, dv/dt]
    
    dxdt = [x[1], -(k / m) * x[0]]
    return dxdt
```

```python
# Initial conditions 1
x01 = [0.1, 2.0]  # Initial displacement and velocity
# k = 3675  
k = 18850  
m = 5000

# Time
t = np.linspace(0, 30, 100)
```

```python
# Maximum Amplitude
u_0 = x01[0]
u_dot_0 = x01[1]
omega_n = np.sqrt(k / m)

u_max = np.sqrt(u_0**2 + (u_dot_0 / omega_n)**2)
print(f"u_max: {u_max}")
```

```python
# Solve the system of differential equations using odeint
solution1 = odeint(ode_func, x01, t, args=(k, m))
u_max_solution1 = np.max(solution1[:, 0])
print(f"u_max: {u_max}")
print(f"u_max_solution1: {u_max_solution1}")
```

```python
# Display the solution
plt.plot(t, solution1[:, 0], label='Undamped Free Vibration')
plt.xlabel('Time (t)')
plt.ylabel('Displacement')
plt.legend()
plt.show()
```

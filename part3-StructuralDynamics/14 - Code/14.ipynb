```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ode_func_mg(x, t, k, m, g, h):
    # x = [displacement, velocity]
    # x[0] = displacement
    # x[1] = velocity
    
    # dxdt = [d(displacement)/dt, d(velocity)/dt]
    # dxdt = [v, dv/dt]
    
    dxdt = [x[1], (g / h - k / m) * x[0]]
    return dxdt
```

```python
def ode_func(x, t, k, m):
    # x = [displacement, velocity]
    # x[0] = displacement
    # x[1] = velocity
    
    # dxdt = [d(displacement)/dt, d(velocity)/dt]
    # dxdt = [v, dv/dt]
    
    dxdt = [x[1], -(k / m) * x[0]]
    return dxdt
```

```python
# Initial conditions 1
x01 = [0.1, 0.0]  # Initial displacement and velocity
# k = 3675  
k = 18850  
m = 5000
g = 9.81
h = 4

# Time
t = np.linspace(0, 30, 100)
```

```python
# Solve the system of differential equations using odeint
solution1 = odeint(ode_func_mg, x01, t, args=(k, m, g, h))
solution2 = odeint(ode_func, x01, t, args=(k, m))
```

```python
# Display the solution
plt.plot(t, solution1[:, 0], label='With gravity effects')
plt.plot(t, solution2[:, 0], label='Without gravity effects')
plt.xlabel('Time (t)')
plt.ylabel('Displacement')
plt.legend()
plt.show()
```
